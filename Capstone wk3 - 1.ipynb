```python
import ssl
import urllib.request, urllib.parse, urllib.error
from urllib.parse import urljoin
from urllib.parse import urlparse
from urllib.parse import quote
import re
from bs4 import BeautifulSoup

import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Not all systems have this so conditionally define parser
try:
    import dateutil.parser as parser
except:
    pass

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
document = urllib.request.urlopen(url, None, 30, context=ctx)
text = document.read()
soup = BeautifulSoup(text, "html.parser")
data = soup.find_all("table")
```

```python
#print to show data of data of table
#[0] to ensure that we get only first data that we need
print(data[0])
wikitable = data[0]
```

```python
#Find out how many tr (row) for this table
print('Number of row = ', len(wikitable.find_all("tr")))
```

```python
wikitable.find_all("tr")[0]
```

```python
#Create table with column name first
neighborhoods = pd.DataFrame(columns=['Postcode', 'Borough', 'Neighbourhood'])
```

```python
#use function to loop for the table and see the result in list
for data in wikitable.find_all("tr"):
   data.find_all('td')
   nn = 1
   row = []
   for data1 in data.find_all('td'):
        data1.text
        #print(data1.text)
        row.append(data1.text.rstrip()) #since last data in each row will contain \n, rtrip will help to remove
   print(row)
```

```python
#put the data from row and use function from above and add to the panda
for data in wikitable.find_all("tr"):
   data.find_all('td')
   nn = 1
   row = []
   for data1 in data.find_all('td'):
        data1.text
        #print(data1.text)
        row.append(data1.text.rstrip())
   try:
      neighborhood_postcode = row[0]
      neighborhood_borough = row[1]
      neighborhood_neighbourhood = row[2]
      neighborhoods = neighborhoods.append({'Postcode': neighborhood_postcode,
                                          'Borough': neighborhood_borough,
                                          'Neighbourhood': neighborhood_neighbourhood}, ignore_index=True)
   except:
      pass
```

```python
neighborhoods
```

```python
nbhassigned = neighborhoods[neighborhoods.Borough != 'Not assigned']
nbhassigned
```

```python
#Find out the Borough with Neighbourhood is 'Not assigned'
nbhassigned[nbhassigned['Neighbourhood'] == 'Not assigned']
```

```python
#Replace row 8 for Neighbourhood data with Borough data
nbhassignedonly = nbhassigned.replace(nbhassigned['Neighbourhood'][8], nbhassigned['Borough'][8])
nbhassignedonly
```

```python
#create table data with the data is duplicated
#nbhassignedonly[['Postcode', 'Borough']][nbhassignedonly[['Postcode', 'Borough']].duplicated() == True]
#nbhassignedonly['Borough'].tolist()
```

```python
#use groupby 'Postcode' and 'Borough' to find duplicate data
#then use aggregate function to combine all duplicate data
groupnbh = nbhassignedonly.groupby(['Postcode', 'Borough'], sort = False).agg(lambda x: ', '.join(x))
groupnbh
```

```python
#However, the index is set to be Postcode and Borough so we will reset index to running number
groupnbhi = groupnbh.reset_index()
groupnbhi
```
